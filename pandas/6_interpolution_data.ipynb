```python
import pandas as pd
```

```python
data = {
    "Name":["Ram","Shyam","Ghanshyam","Krishna","Arjun","Karna","Sita","Radh"],
    "Age":[10,None,30,40,50,60,70,80],
    "Salary":[50000,None,70000,80000,90000,100000,110000,120000],
    "City":["Nagpur","Mumbai","Delhi","Pune","Bangalore","Kolkata","Chennai","Ahmedabad"],
    "Performance Score":[90,None,65,76,87,34,76,100]
}

df = pd.DataFrame(data)
df
```

```python
# interpolution (missing value find and estimated value fill)
# use to interpolution 
# - preserve data integraity
# - smooth trends
# method => linear,polynomial,time
# df['Age'].interpolate(method='linear',axis=0,inplace=True) # single column data interpolate
df.interpolate(method='linear',axis=0,inplace=True) # all table data interpolate
df
```
