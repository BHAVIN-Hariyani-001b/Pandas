- how to select specific column

```python
# select column 
print(df['Name']) 
df[df['Age'] > 52] # apply bool condition 
```

```python
# select mutiple column 
df[['Name','City']]
```

```python
# filter rows
high_salary = df[df['Salary'] > 80000]
high_salary
```

```python
# mutiple condition use and filter data
# using AND (&) 
df[(df['Salary'] > 80000) & (df['Age'] > 40)]
```

```python
# using OR (|) 
df[(df['Age'] > 50) | (df['Performance Score'] > 90)]
```
