```python
import pandas as pd
```

```python
data = {
    "Name":["Ram","Shyam","Ghanshyam","Krishna","Arjun","Karna","Sita","Radh"],
    "Age":[10,20,30,40,50,60,70,80],
    "Salary":[50000,60000,70000,80000,90000,100000,110000,120000],
    "City":["Nagpur","Mumbai","Delhi","Pune","Bangalore","Kolkata","Chennai","Ahmedabad"],
    "Performance Score":[90,98,65,76,87,34,76,100]
}

df = pd.DataFrame(data)
df
```

```python
# data sort 
# column name pass 
# data is sort to ascending/descending you can pass true for ascending and false for descending
df.sort_values(by="Performance Score",ascending=True,inplace=True)
df
```

```python
df.sort_values(by="Performance Score",ascending=False,inplace=True)
df
```

```python
#mutiple row sort and pass to list
df.sort_values(by=["Age","Salary"],ascending=[True,False],inplace=True) 
df
```
