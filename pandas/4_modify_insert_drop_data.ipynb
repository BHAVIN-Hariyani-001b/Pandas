```python
import pandas as pd
```

```python
data = {
    "Name":["Ram","Shyam","Ghanshyam","Krishna","Arjun","Karna","Sita","Radh"],
    "Age":[10,20,30,40,50,60,70,80],
    "Salary":[50000,60000,70000,80000,90000,100000,110000,120000],
    "City":["Nagpur","Mumbai","Delhi","Pune","Bangalore","Kolkata","Chennai","Ahmedabad"],
    "Performance Score":[90,34,65,76,87,34,76,100]
}

df = pd.DataFrame(data)
df
```

```python

# add column
# column is always insert in last
df["Bonus"] = df['Salary'] * 10 / 100  # add column and find 10% of salary and put in new column
df
```

```python
# useing insert method add column
df.insert(0,"Emp id",[1,2,3,4,5,6,7,8])
df
```

```python
# modify data specific row(index) and  column(name)
df.loc[0,'Salary'] = 55000
df
```

```python
# increse salary by 5% all data
# df['Salary'] = df['Salary'] + (df['Salary'] * 5 / 100)
## OR
df['Salary'] = df['Salary'] * 1.05
df
```

```python
# drop method is use to delete column or row
df.drop(columns=['Bonus'],inplace=True) # inplace is use to all data remove to bonus column
df
```

```python
# mutiple column data remove
df.drop(columns=['Emp id','Age'],inplace=True)
df
```
