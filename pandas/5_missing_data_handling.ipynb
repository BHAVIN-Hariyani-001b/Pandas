```python
import pandas as pd
```

```python
# missing data is like show on like {NaN} OR {None}

# isnull() function halp to use and return true or false
# True -- Nan is missing
# False -- value is present
```

```python
data = {
    "Name":["Ram",None,"Ghanshyam","Krishna","Arjun","Karna","Sita","Radh"],
    "Age":[10,None,30,40,50,60,70,80],
    "Salary":[50000,None,70000,80000,90000,100000,110000,120000],
    "City":["Nagpur",None,"Delhi","Pune","Bangalore","Kolkata","Chennai","Ahmedabad"],
    "Performance Score":[90,None,65,76,87,34,76,100]
}

df = pd.DataFrame(data)
print(df,end='\n\n')
print(df.isnull()) # detect missing data
df.isnull().sum() # find how many missing value in dataframe and return count of number
```

```python
# handling missing data (delete missing data)
# df.dropna(axis=0,inplace=True)
df
```

```python
# find and fill the missing data
# df['Age'].fillna(df['Age'].mean(),inplace=True)
## OR
df['Age'] = df['Age'].fillna(df['Age'].mean()).round(2)
df['Salary'] = df['Salary'].fillna(df['Salary'].mean()).round(2)
df['Performance Score'] = df['Performance Score'].fillna(df['Performance Score'].mean()).round(2)
df['Name'] = df['Name'].fillna('xyz')
df['City'] = df['City'].fillna('abc')

# df['Age'] = df['Age'].round(2) # age is mutiple zero after point and round() function use and decrese to zero
# df.fillna(0,inplace=True) ## nan and none value find allow value fill is zero
df
```
