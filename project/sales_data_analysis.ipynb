```python
import pandas as pd
```

```python
df = pd.read_csv("./data_file/sales_data.csv")
df
```

```python
### Add Sales Amount (Quantity * Price)
df["SalesAmount"] = df["Price"] * df["Quantity"]
df
```

```python
### Total sale's
total_sale = df["SalesAmount"].sum()
total_sale
```

```python
### Average Order Value
oreder_avg = df["SalesAmount"].mean()
oreder_avg
```

```python
### Top Customer by Sales`s
top_customer = df.groupby("Customer")["SalesAmount"].sum().idxmax()
top_customer
```

```python
### Top Product by Sales
top_product = df.groupby("Product")["SalesAmount"].sum().idxmax()
top_product
```

```python
### Region-wise total sales
region_sales = df.pivot_table(values="SalesAmount",index="Region",aggfunc=["sum"])
region_sales
```

```python
### Category-wise total & average sales
category_sales = df.pivot_table(values="SalesAmount",index="Category",aggfunc=["sum","mean"])
category_sales
```

```python
### Top product per category (like subject-wise topper)
product_sales = df.groupby(["Category","Product"])["SalesAmount"].sum().reset_index()
category_topper = product_sales.sort_values(["Category","SalesAmount"],ascending=[True,False]).drop_duplicates("Category")
category_topper
```
