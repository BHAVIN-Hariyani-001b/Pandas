```python
import pandas as pd
```

```python
customer = pd.read_csv("./data_file/customers.csv")
oreder = pd.read_csv("./data_file/orders.csv")
product = pd.read_csv("./data_file/products.csv")
```

```python
### Add Revenue column
df = oreder.merge(customer,on="CustomerID",how="left") \
    .merge(product,on="ProductID",how="left")
df
```

```python
df["Revenue"] = df["Quantity"] * df["Price"]
df
```

```python
###  Most purchased product
most_purchased = df.groupby("Product")["Quantity"].sum().idxmax()
most_purchased
```

```python
### Revenue per customer
rev_per_customer = df.groupby(["Name","CustomerID"])["Revenue"].sum().reset_index()
rev_per_customer
```

```python
### Top revenue customer
top_rev_customer = rev_per_customer.loc[rev_per_customer["Revenue"].idxmax()]
top_rev_customer
```
